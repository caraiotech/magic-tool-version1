```python
import pickle
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def padded_chromosome(chromosome,dim):
    h,w,_=chromosome.shape
    print(h,w)
    h_add=dim[0]-h
    w_add=dim[1]-w
    top=h_add//2
    bottom=h_add//2
    left=w_add//2
    right=w_add//2
    white = [255,255,255]
    padded_img = cv2.copyMakeBorder(chromosome, top, bottom, left, right,cv2.BORDER_CONSTANT,value=white)
    return padded_img
```

```python
with open(r"C:\Users\HP\Desktop\CARAIOLocal\prototype_fin\magic-tool-prototype\app\chr_85_62.pkl","rb") as f:
    img=pickle.load(f)
dim=(85,62)
plt.imshow(img)
plt.show()
padded_chromosome(img,dim)
```
